```python
def unidade_responsavel (df, codigo, nivel):
    return df[(df.codigo==codigo) & (df.nivel==nivel)].unidade_responsavel.values[0]
def nome (df, codigo, nivel):
    return df[(df.codigo==codigo) & (df.nivel==nivel)].nome.values[0]
def ch_total (df, codigo, nivel):
    return df[(df.codigo==codigo) & (df.nivel==nivel)].ch_total.values[0]
def tipo_componente (df, codigo, nivel):
    return df[(df.codigo==codigo) & (df.nivel==nivel)].tipo_componente.values[0]
def modalidade (df, codigo, nivel):
    return df[(df.codigo==codigo) & (df.nivel==nivel)].modalidade.values[0]
def ementa (df, codigo, nivel):
    return df[(df.codigo==codigo) & (df.nivel==nivel)].ementa.values[0]
def objetivos (df, codigo, nivel):
    return df[(df.codigo==codigo) & (df.nivel==nivel)].objetivos.values[0]
def conteudo (df, codigo, nivel):
    return df[(df.codigo==codigo) & (df.nivel==nivel)].conteudo.values[0]
def competencias_habilidades (df, codigo, nivel):
    return df[(df.codigo==codigo) & (df.nivel==nivel)].competencias_habilidades.values[0]
def referencias (df, codigo, nivel):
    return df[(df.codigo==codigo) & (df.nivel==nivel)].referencias.values[0]
```
